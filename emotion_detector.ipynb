```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import IPython as ipad
import librosa
import librosa.display as ld
import os
```

```python
paths = []  # To store the file paths
labels = []  # To store the labels extracted from folder names
```

```python
folder_path= 'emotion_detector/TESS'
```

```python
folder_path = 'TESS'


# Walk through the folder
for dirname, _, filenames in os.walk(folder_path):
    for filename in filenames:
        paths.append(os.path.join(dirname, filename))
        label = filename.split('_')[-1]
        label = label.split('.')[0]
        labels.append(label.lower())
```

```python
labels
```

```python
df=pd.DataFrame()
df["path"]=paths
df["label"]=labels
```

```python
print(df)
```

```python
def feature_extraction(file):
    audio,sample_rate=librosa.load(file)
    mfscc=librosa.feature.mfcc(y=audio,sr=sample_rate,n_mfcc=40)
    scaled_feature=np.mean(mfscc.T,axis=0)
    return scaled_feature
```

```python
data=[]
```

```python
df["data"]=df["path"].apply(feature_extraction)
```

```python

from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
y = enc.fit_transform(df[['label']])
print(y.shape)
```

```python

x = np.array(df["data"].tolist())  # Convert list of arrays to a 2D numpy array
x = np.expand_dims(x, -1) 
```

```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Flatten,Dropout
from keras.models import Sequential
```

```python
print(y.shape)
```

```python

y = y.toarray()

```

```python

from tensorflow.keras.layers import LSTM
```

```python
models = Sequential ([
    LSTM(123, return_sequences=False, input_shape=(40,1)),
    Dense(64, activation='relu'),
    Dropout(0,2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(7, activation='softmax'),
])

models.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
models.summary()
```

```python
history = models.fit(x, y, validation_split=0.2, epochs=100, batch_size=512, shuffle=True)
```

```python
def record_audio(duration, fs=22050):
    print("Recording...")
    audio = sd.rec(int(duration * fs), samplerate=fs, channels=1, dtype='float32')
    sd.wait()  # Wait until the recording is finished
    print("Recording complete.")
    return np.squeeze(audio)
```

```python
def max_repeating_element(arr):
    # Create a dictionary to store the frequency of each element
    freq = {}
    
    # Iterate over the array and count the occurrences of each element
    for elem in arr:
        if elem in freq:
            freq[elem] += 1
        else:
            freq[elem] = 1
    
    # Find the element with the maximum frequency
    max_elem = max(freq, key=freq.get)
    
    return max_elem, freq[max_elem]
```
